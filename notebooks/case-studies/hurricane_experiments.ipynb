```python
import hurdat2parser as hd2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
atl = hd2.Hurdat2(basin="atl")
```

```python
katrina_track = np.array(atl[2005]["KATRINA"].gps)
sandy_track = np.array(atl["_SANDY"].gps)
plt.plot(katrina_track[:, 1], katrina_track[:, 0], marker="o", color="red", label="Katrina")
plt.plot(sandy_track[:, 1], sandy_track[:, 0], marker="o", color="blue", label="Sandy")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend()
```
